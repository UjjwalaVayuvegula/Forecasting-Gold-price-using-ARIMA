```python
aggregated_data = data1.groupby(['CustomerID','City','DateOfRegistration','OnlineCommunication','AutomaticRefill','DoorstepDelivery','PreferredDeliveryDay']).agg({'Dateofemail': lambda x: ', '.join(x),'EmailType': lambda x: ', '.join(x),'timestamp_of_sent': lambda x: ', '.join(x),'MailOpened': lambda x: ', '.join(x),
                                                                                                                                                                           'MailClicked': lambda x: ', '.join(x),'DateofOrder': lambda x: ', '.join(x),
                                                                                                                                                                           'timestamp_of_opening': lambda x: ', '.join(x),'RetentionPotential': lambda x: ', '.join(x),'OrderQuantity':'sum'}).reset_index()
```

```python
Dateofemail	= set()
EmailType = set()
timestamp_of_sent = set()
MailOpened = set()
MailClicked = set()
DateofOrder = set()
timestamp_of_opening = set()
RetentionPotential = set()

for row in aggregated_data.itertuples():
    Dateofemail.update(row.Dateofemail.split(', '))
    EmailType.update(row.EmailType.split(', '))
    timestamp_of_sent.update(row.timestamp_of_sent.split(', '))
    MailOpened.update(row.MailOpened.split(', '))
    MailClicked.update(row.MailClicked.split(', '))
    DateofOrder.update(row.DateofOrder.split(', '))
    timestamp_of_opening.update(row.timestamp_of_opening.split(', '))
    RetentionPotential.update(row.RetentionPotential.split(', '))

# Create new columns for each email, timestamp, and date
for date in Dateofemail:
    aggregated_data[date] = aggregated_data['Dateofemail'].str.contains(date).astype(int)
for email in EmailType:
    aggregated_data[email] = aggregated_data['EmailType'].str.contains(email).astype(int)
for timestampsent in timestamp_of_sent:
    aggregated_data[timestampsent] = aggregated_data['timestamp_of_sent'].str.contains(timestampsent).astype(int)
for open in MailOpened:
    aggregated_data[open] = aggregated_data['MailOpened'].str.contains(open).astype(int)
for click in MailClicked:
    aggregated_data[click] = aggregated_data['MailClicked'].str.contains(click).astype(int)
for D in DateofOrder:
    aggregated_data[D] = aggregated_data['DateofOrder'].str.contains(D).astype(int)
for time in timestamp_of_opening:
    aggregated_data[time] = aggregated_data['timestamp_of_opening'].str.contains(time).astype(int)
for P in RetentionPotential:
    aggregated_data[P] = aggregated_data['RetentionPotential'].str.contains(P).astype(int)

# Drop the original columns containing the lists
aggregated_data = aggregated_data.drop(['Dateofemail', 'EmailType', 'timestamp_of_sent','MailOpened','MailClicked','DateofOrder','timestamp_of_opening','RetentionPotential'], axis=1)
```
